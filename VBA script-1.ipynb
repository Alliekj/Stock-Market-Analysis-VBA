```python
Sub project()
    Dim row As Long
    Dim Output_Row As Long
    Dim This_ticker As String
    Dim Next_Ticker As String
    Dim OpenValue As Double
    Dim CloseValue As Double
    Dim Volume As LongLong
    Dim Total_Volume As LongLong
    Dim NumberRows As Long
    Dim This_Sheet As Worksheet
    
        For Each This_Sheet In Worksheets
            This_Sheet.Select
         
         NumberRows = ActiveSheet.UsedRange.Rows.Count
        
        Output_Row = 1
        Cells(Output_Row, 9).Value = "Ticker"
                Cells(Output_Row, 10).Value = "Yearly Change"
                Cells(Output_Row, 11).Value = "Percent Change"
                Cells(Output_Row, 12).Value = "Total Stock Volume"
                Output_Row = Output_Row + 1
        
        Total_Volume = 0
        For row = 2 To NumberRows
        
        'Debug.Print row
            Prev_Ticker = Cells(row - 1, 1).Value
            This_ticker = Cells(row, 1).Value
            Next_Ticker = Cells(row + 1, 1).Value
             Volume = Cells(row, 7).Value
             Total_Volume = Total_Volume + Volume
             
             
             If Prev_Ticker <> This_ticker Then
                    Open_Value = Cells(row, 3).Value
            ElseIf Next_Ticker <> This_ticker Then
                Close_Value = Cells(row, 6).Value
                Cells(Output_Row, 9).Value = This_ticker
                
                Cells(Output_Row, 10).Value = Close_Value - Open_Value
                 Cells(Output_Row, 10).NumberFormat = "0.00"
                If Close_Value - Open_Value > 0 Then
                    Cells(Output_Row, 10).Interior.Color = RGB(0, 255, 0)
                Else
               Cells(Output_Row, 10).Interior.Color = RGB(255, 0, 0)
                End If
                Cells(Output_Row, 11).Value = (Close_Value - Open_Value) / Open_Value
                Cells(Output_Row, 11).NumberFormat = "0.00%"
               Cells(Output_Row, 12).Value = Total_Volume
               Cells(Output_Row, 12).NumberFormat = "#,###"
                Total_Volume = 0
                
                
                
                Output_Row = Output_Row + 1
            End If
            'Debug.Print row, ticker
        
           
        Next row
        
        Next This_Sheet
    

End Sub

```
